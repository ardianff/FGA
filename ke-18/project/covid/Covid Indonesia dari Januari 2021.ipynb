```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("covid_19_data.csv")
```

```python
indonesia = df[df['Country/Region'] == "Indonesia"]
indonesia
```

```python
kasus = indonesia[indonesia['Confirmed']>0]
kasus['ObservationDate'] = pd.to_datetime(kasus['ObservationDate'])
kasus
```

```python
plt.plot(kasus.ObservationDate, kasus.Confirmed)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
kasusjan = kasus[kasus.ObservationDate>='1/1/2021']
kasusjan
```

```python
plt.plot(kasusjan.ObservationDate, kasusjan.Confirmed, color='blue')
plt.plot(kasusjan.ObservationDate,kasusjan.Recovered, color='green')
plt.plot(kasusjan.ObservationDate,kasusjan.Deaths, color='red')
plt.xticks(rotation = 'vertical')
plt.legend(['Jumlah Kasus','Jumlah Sembuh','Jumlah Meninggal'])
plt.title('Inforgrafis Covid-19 Di Indonesia Mulai Januari 2021')
plt.show()
```
