```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("covid_19_data.csv")
```

```python
kasus1 = df[df['Country/Region'] == 'Indonesia']
kasus2 = df[df['Country/Region'] == 'Malaysia']
kasus1
```

```python
kasusind = kasus1[kasus1['Confirmed']>0]
kasusmly = kasus2[kasus2['Confirmed']>0]
kasusind.head(10)
```

```python
kasusmly.head(10)
```

```python
kasusind['ObservationDate'] = pd.to_datetime(kasusind['ObservationDate'])
kasusind.head()
```

```python
kasusmly['ObservationDate'] = pd.to_datetime(kasusmly['ObservationDate'])
kasusmly.head()
```

```python
kasusind['BaseDate'] = pd.to_datetime('2020-03-02')
kasusmly['BaseDate'] = pd.to_datetime('2020-01-25')
kasusmly.head()
```

```python
kasusind.head()
```

```python
kasusind['HariKe'] = kasusind['ObservationDate'] - kasusind['BaseDate']
kasusmly['HariKe'] = kasusmly['ObservationDate'] - kasusmly['BaseDate']
```

```python
kasusind.head()
```

```python
kasusind.HariKe = kasusind.HariKe/pd.Timedelta(1, unit='d')
kasusind.head()
```

```python
kasusmly.HariKe = kasusmly.HariKe/pd.Timedelta(1, unit='d')
kasusmly.head()
```

```python
ind=plt.bar(kasusind.HariKe, kasusind.Confirmed)
mly=plt.bar(kasusmly.HariKe, kasusmly.Confirmed)

plt.xlabel('Hari ke-')
plt.ylabel('Jumlah Kasus')
plt.legend(['Indonesia','Malaysia'])
plt.title('Jumlah Perbandingan Kasus Covid-19')
plt.show()
```
